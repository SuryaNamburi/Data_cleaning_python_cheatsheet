# Step : 4 Standardize Data Formats

### 4.1__Convert to lower/upper case

```python
df["column_name"] = df["column_name"].str.lower()
df["column_name"] = df["column_name"].str.upper()
```

### 4.2__Formate Dates

```python
df["column_name"] = pd.to_datetime(df["column_name"])
```

### 4.3__Convert data Type

```python
df["string_colum"] = df["string_colum"].astype("categorical_column")
df["numerical_column"] = df["numerical_column"].astype("float")
```

# Step : 5 Remove Duplicates

```python
df = df.drop_duplicates()
```

# Step : 6 Handling Missing Values

### 6.1__Fill with O (when O is meaningful)

```python
df["column_name"] = df["column_name"].fillna(0)
```

### 6.2__Drop rows if missing data is critical

```python
df = df[df["column_name"].notna(]
```

### 6.3__Fill with average (numerical columns only)

```python
df["numeric_colum"] = df["numeric_column"].fillna(df["numeric_column"].mean())
```
